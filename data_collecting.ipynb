### Get 2 songs for each singer in top 100 and save to JSON

```python
songs = {}
for i, singer in top100.iterrows():
    if (((i+1)%30)==0):
        print('*'*10,' Idling', '*'*10)
        time.sleep(30)
    else:
        artist_song = artist.Artist(singer.artist).get_songs(results=2)
        for j, song in enumerate(artist_song):
            songs[song.id] = song.title
            print('{0}/60|{1}'.decode("iso-8859-9").format((i*2+j)%60,song.title))
os.system("say 'hello world again'")
```

```python
with open('songs.json', 'w') as outfile:
    json.dump(songs, outfile)
```
