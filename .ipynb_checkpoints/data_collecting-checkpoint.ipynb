### Each Singer is a class project. So we store them with JSON

```python
def artist2json(singer):
    result = {'id':singer.id,
              'name':singer.name,
              'hotttnesss':singer.hotttnesss,
              'similar': [s.name for s in singer.similar]
             }
    return result
```

```python
artist_json = artist2json(hottt_artist[0])
print artist_json
```

### Try to store Top 50 into one JSON

```python
# merge all singer names into header
with open('Top50.json', 'w') as outfile:
    header = [artist2json(singer) for singer in hottt_artist]
    json.dump(header, outfile)
```

```python
header = json.load(open('Top50.json'))
```

```python
artist_list = [] 
similarity = []
for singer in header:
    artist_list.append(singer['name'])
    similarity.append(singer['name'])
    for similar in singer['similar']:
        similarity.append(singer['name'])
        artist_list.append(similar)
```

```python
artist_set = list(set(artist_list))
```

```python
artist_array = np.array(artist_set)
```

### create array to store the hotness

```python
artist_hottt = np.zeros(len(artist_array))
```

```python
# decode("iso-8859-9") will solve the problem of print European names that not in utf8.
# with the trick of sleeping per 60s to avoid error
# with the trick of say "hello world" to inform you the finishing of the work
import time
for i, singer in enumerate(artist_array):
    if (((i+1)%60)==0):
        print('*'*10,' Idling', '*'*10)
        time.sleep(60)
    else:
        artist_hottt[i] = artist.Artist(singer).hotttnesss
    print('{0}/60|{1}|{2}'.encode("iso-8859-9").format((i+1)%60,singer,artist_hottt[i]))
os.system("say 'hello world'") 
```

```python
print len(similarity)
print len(artist_list)
```

### Now create csv to save the names and score

```python
data = pd.DataFrame({'artist':artist_array,
                    'hotttnesss':artist_hottt})
data = data.sort_values(by='hotttnesss',ascending=0).reset_index(drop=True)
data.to_csv('similar_artists_top_50.csv',encoding="iso-8859-9",index=False)
```

```python
top100 = pd.read_csv('similar_artists_top_50.csv', encoding="iso-8859-9", nrows=100)
```

### Get 2 songs for each singer in top 100 and save to JSON

```python
songs = {}
for i, singer in top100.iterrows():
    if (((i+1)%30)==0):
        print('*'*10,' Idling', '*'*10)
        time.sleep(30)
    else:
        artist_song = artist.Artist(singer.artist).get_songs(results=2)
        for j, song in enumerate(artist_song):
            songs[song.id] = song.title
            print('{0}/60|{1}'.decode("iso-8859-9").format((i*2+j)%60,song.title))
os.system("say 'hello world again'")
```

```python
with open('songs.json', 'w') as outfile:
    json.dump(songs, outfile)
```
